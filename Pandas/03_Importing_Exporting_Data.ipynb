## Section 1: Reading CSV

The most common way to load data into Pandas is from a CSV file using `pd.read_csv()`.

While the basic usage is simple, `read_csv()` has powerful parameters that let you control exactly how the data is loaded.

### Key parameters covered:

| Parameter | What it does |
|---|---|
| `usecols` | Load only specific columns — saves memory on large files |
| `nrows` | Load only the first N rows — useful for previewing large files |
| `na_values` | Treat custom strings as NaN at import time |
| `index_col` | Set a column as the DataFrame index on load |

````python
# Basic Read
df = pd.read_csv("AB_NYC_2019.csv")
df.head(3)
````

````python
# read_csv with key parameters
df = pd.read_csv(
    "AB_NYC_2019.csv",
    usecols=["name", "neighbourhood_group", "room_type", "price", "minimum_nights", "reviews_per_month"],
    nrows=100
)
df.head(3)
````

````python
# Handling missing values at import time
df = pd.read_csv(
    "AB_NYC_2019.csv",
    na_values=["NA", "N/A", "none", ""]
)
df.head(3)
````

````python
# Setting a column as index at import time
df = pd.read_csv(
    "AB_NYC_2019.csv",
    index_col="id"
)
df.head(3)
````

## Section 3. Reading JSON 

#### What is JSON?
JSON (JavaScript Object Notation) is a lightweight data interchange format that's easy for humans to read and write and easy for machines to parse and generate. It's language-independent but uses conventions familiar to programmers of C-family languages.

#### Key characteristics:

- Text-based format
- Self-describing
- Hierarchical structure
- Commonly used for transmitting data in web applications

#### How JSON Looks Like

JSON data is written as key/value pairs

 Here are examples:

### JSON Structure — Key Concepts

JSON data is built from two structures:
- **Objects** — key/value pairs wrapped in `{}` (like a Python dictionary)
- **Arrays** — ordered lists wrapped in `[]` (like a Python list)

These can be **nested** inside each other to represent hierarchical data.

### The `orient` parameter
`orient` tells Pandas how the JSON is structured:

| orient | Structure |
|---|---|
| `"records"` | List of row objects `[{col: val}, ...]` — most common |
| `"columns"` | Dict of column arrays `{col: {index: val}}` |
| `"index"` | Dict of row dicts `{index: {col: val}}` |
| `"values"` | Just a 2D array of values |

````python
# Basic JSON object
{
  "name": "John Doe",
  "age": 30,
  "city": "New York",
  "isStudent": False
}
````

````python
# JSON Array
{
  "employees": [
    {"firstName": "John", "lastName": "Doe"},
    {"firstName": "Anna", "lastName": "Smith"},
    {"firstName": "Peter", "lastName": "Jones"}
  ]
}
````

````python
# Nested JSON
{
  "library": "City Library",
  "books": [
    {
      "title": "Python Basics",
      "author": {
        "name": "Jane Smith",
        "email": "jane@example.com"
      },
      "year": 2020,
      "available": True
    }
  ]
}
````

## Section 5: Exporting Data

Once we've loaded, cleaned, or transformed our data, we'll often need to export it — to share with teammates, feed into another tool, or save a cleaned version.

Pandas makes this straightforward with `.to_csv()`, `.to_excel()`, and `.to_json()`.

#### Key parameter — `index=False`

By default, Pandas writes the DataFrame index as an extra column in the output.

Almost always we want `index=False` to suppress this — otherwise we'll get an unnamed column of row numbers when the file is reopened.

#### Export methods summary

| Method | Output | Key parameter |
|---|---|---|
| `df.to_csv("file.csv", index=False)` | CSV file | `index`, `sep`, `encoding` |
| `df.to_excel("file.xlsx", sheet_name="Sheet1", index=False)` | Excel file | `sheet_name`, `index` |
| `df.to_json("file.json", orient="records")` | JSON file | `orient` |

````python
# Export to CSV
df.to_csv("airbnb_export.csv", index=False)
````

````python
# Export specific columns only
df[["name", "neighbourhood_group", "room_type", "price"]].to_csv(
    "airbnb_subset.csv",
    index=False
)
````

````python
# Export to Excel with sheet name
df.to_excel(
    "airbnb_export.xlsx",
    sheet_name="NYC Listings",
    index=False
)
````

````python
# Export to JSON
df.to_json("airbnb_export.json", orient="records")
````
